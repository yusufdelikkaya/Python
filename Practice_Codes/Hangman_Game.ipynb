**Senaryo ve Talimatlar**

**Senaryo**

Bir grup arkadaş, birlikte eğlenmek ve kelime tahmin yeteneklerini geliştirmek için basit bir adam asmaca oyunu oynamak istiyor. Oyun, rastgele bir kelime seçer ve oyuncuların bu kelimeyi harf harf tahmin etmelerini sağlar. Her yanlış tahminde oyuncunun tahmin hakkı azalır ve tüm hakları bittiğinde oyun sona erer.

**Talimatlar**

1. Amaç

Bu adam asmaca oyunu, oyuncuların rastgele seçilen bir kelimeyi harf harf tahmin ederek bulmalarını sağlar. Oyuncular, her seferinde bir harf tahmin eder ve doğru tahminler yaparak kelimeyi bulmaya çalışırlar.

4. Kullanım Talimatları

    Oyuna Başla: Menüden 1'i seçin ve oyuna başlayın.
    Harf Tahmin Et: Her seferinde bir harf tahmin edin. Doğru tahminler kelimeyi ortaya çıkarır, yanlış tahminler tahmin hakkınızı azaltır.
    Çık: Menüden 2'yi seçin ve programdan çıkın.

````python
# Adım 2: game.py

# Bu modül, oyun mantığını ve gerekli fonksiyonları içerecek.

import random

def choose_word():
    words = [
        "python",
        "programming",
        "developer",
        "software",
        "algorithm",
        "function",
        "variable",
        "iteration",
        "comprehension",
        "nested"
    ]
    return random.choice(words)

def display_word(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def get_guess():
    guess = input("Bir harf tahmin edin: ").lower()
    return guess

def play_game():
    word = choose_word()
    guessed_letters = set()
    attempts_left = 6

    print("Adam Asmaca Oyununa Hoşgeldiniz!")
    print("Kelimeyi tahmin etmek için harfleri girin. Toplamda 6 hakkınız var.")

    while attempts_left > 0:
        print(f"\nKalan Tahmin Hakkı: {attempts_left}")
        print(display_word(word, guessed_letters))

        guess = get_guess()

        if guess in guessed_letters:
            print("Bu harfi zaten tahmin ettiniz.")
        elif guess in word:
            guessed_letters.add(guess)
            print("Doğru tahmin!")
            if all(letter in guessed_letters for letter in word):
                print(f"Tebrikler! Kelimeyi buldunuz: {word}")
                break
        else:
            attempts_left -= 1
            print("Yanlış tahmin.")

        if attempts_left == 0:
            print(f"Üzgünüm, tahmin hakkınız bitti. Kelime: {word}")
````

````python
# Adım 3: main.py

# Bu modül, oyunu başlatan ana programı içerecek.

from game import play_game

def display_menu():
    print("\nAdam Asmaca Oyunu")
    print("1. Oyuna Başla")
    print("2. Çık")

def main():
    while True:
        display_menu()
        choice = input("Bir seçenek seçin (1-2): ")

        if choice == "1":
            play_game()
        elif choice == "2":
            print("Çıkılıyor...")
            break
        else:
            print("Geçersiz seçenek. Lütfen tekrar deneyin.")

if __name__ == "__main__":
    main()
    

# bu dosya çalıştırılacak
````
